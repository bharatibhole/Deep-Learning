```
import numpy as np
x = np.linspace(-2, 4, 20)   
plt.plot(x, np.sin(x),'.','r',linewidth=3) 
plt.plot(x, np.cos(x),'b-',linewidth=3) 
plt.title("Plot $x^2$")
plt.annotate("local maxima", xy=(2, -1))
plt.xlim(-3,5)
plt.xlim(-3,5)
plt.show()
```
